```python
import numpy as np
a = np.linspace(0,1,24)
b = np.random.rand(4,6)
c = np.sort(b, axis=None).reshape(4,6)
add_arr = np.array([0.69, 0.96, -0.69, 0.22, 0.67, -0.76])
c = c + add_arr
print(c)
c = c.reshape(3,8)
c = c.T
print(c)
mean_value = c.mean()
print(mean_value)
locations = np.where(c > mean_value)
print(locations)

```
