# An MDP is defined by:
# - States
# - Actions
# - Transition probabilities
# - Rewards

# We will represent transitions using dictionaries for simplicity


```python
class MDP:
    def __init__(self, states, actions, transitions, gamma):
        self.states = states
        self.actions = actions
        self.transitions = transitions
        self.gamma = gamma

    def get_actions(self, state):
        return self.actions[state]

    def get_transitions(self, state, action):
        return self.transitions[state][action]
```

```python
states = [0, 1, 2]
actions = {
    0: [0, 1],
    1: [0, 1],
    2: [0, 1]
}

# transitions[state][action] = list of (prob, next_state, reward)
transitions = {
    0: {
        0: [(1.0, 0, 0)],
        1: [(1.0, 1, 1)]
    },
    1: {
        0: [(1.0, 0, 0)],
        1: [(1.0, 2, 2)]
    },
    2: {
        0: [(1.0, 2, 0)],
        1: [(1.0, 2, 0)]
    }
}

gamma = 0.9
mdp = MDP(states, actions, transitions, gamma)
```

```python
class ValueIterationAlgorithm:
    def __init__(self, mdp, theta=1e-6):
        self.mdp = mdp
        self.theta = theta
        self.V = {s: 0 for s in mdp.states}
        self.policy = {s: None for s in mdp.states}

    def run_algorithm(self):
        while True:
            delta = 0
            for s in self.mdp.states:
                v_old = self.V[s]
                
                # TODO: compute value for all actions
                action_values = []
                for a in self.mdp.get_actions(s):
                    value = 0
                    for prob, next_state, reward in self.mdp.get_transitions(s, a):
                        value += prob * (reward + self.mdp.gamma * self.V[next_state])
                    action_values.append(value)
                
                self.V[s] = max(action_values)
                delta = max(delta, abs(v_old - self.V[s]))
            
            if delta < self.theta:
                break
```

```python
   class ValueIterationAlgorithm:
    def __init__(self, mdp, theta=1e-6):
        self.mdp = mdp
        self.theta = theta
        self.V = {s: 0 for s in mdp.states}
        self.policy = {s: None for s in mdp.states}

    def run_algorithm(self):
        while True:
            delta = 0
            for s in self.mdp.states:
                v_old = self.V[s]

                action_values = []
                for a in self.mdp.get_actions(s):
                    value = 0
                    for prob, next_state, reward in self.mdp.get_transitions(s, a):
                        value += prob * (reward + self.mdp.gamma * self.V[next_state])
                    action_values.append(value)

                self.V[s] = max(action_values)
                delta = max(delta, abs(v_old - self.V[s]))

            if delta < self.theta:
                break

    def extract_policy(self):   # â† MUST be indented at same level as run_algorithm
        for s in self.mdp.states:
            action_values = {}
            for a in self.mdp.get_actions(s):
                value = 0
                for prob, next_state, reward in self.mdp.get_transitions(s, a):
                    value += prob * (reward + self.mdp.gamma * self.V[next_state])
                action_values[a] = value

            self.policy[s] = max(action_values, key=action_values.get)
 
```

```python
vi = ValueIterationAlgorithm(mdp)
vi.run_algorithm()
vi.extract_policy()

print("Optimal Value Function:")
for s, v in vi.V.items():
    print(f"State {s}: {v:.2f}")

print("\nOptimal Policy:")
for s, a in vi.policy.items():
    print(f"State {s}: Action {a}")
```
